```python
import matplotlib.pyplot as plt
import pandas as pd
Marks={"Arnab":[80,85,94,86,90],"Sakshi":[90,82,79,97,94],
      "Ravi":[96,74,83,97,98],"Mukesh":[76,91,84,76,84],
      "Ajay":[73,94,78,82,81]}
Df=pd.DataFrame(Marks,index=["Physics","Chemistry","Maths","English","IP"])
print(Df)
Df.plot(kind="bar", edgecolor="Black")
plt.title("Result Analysis")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend(loc="lower right")
```

```python
import numpy as np
import matplotlib.pyplot as plt
discount= np.array([10,20,30,40,50])
saleInRs=np.array([40000,45000,48000,50000,100000])
plt.scatter(x=discount,y=saleInRs)
plt.title('Sales Vs Discount')
plt.xlabel('Discount offered')
plt.ylabel('Sales in Rs')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
discount= np.array([10,20,30,40,50])
saleInRs=np.array([40000,45000,48000,50000,100000])
size=discount*10
plt.scatter(x=saleInRs,y=discount,s=size,color='red',linewidth=3,marker='*',edgecolor='blue')
plt.title('Sales Vs Discount')
plt.xlabel('Discount offered')
plt.ylabel('Sales in Rs')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

discount= np.array([10,20,30,40,50])
saleInRs=np.array([40000,45000,48000,50000,100000])
size=discount*10

plt.scatter(x=discount,y=saleInRs,s=size,color='red',linewidth=3,marker='*',edgecolor='blue')
plt.title('Sales Vs Discount')
plt.xlabel('Discount offered')
plt.ylabel('Sales in Rs')
plt.savefig('Scatter.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
Df=pd.DataFrame({"Air": [78,0.9,21,0.1]}, index= ["Nitrogen","Argon","Oxygen","Other Gases"])

Df.plot(kind="pie", y="Air", title="Composition of Air", legend=False, explode=[0.1,0.1,0,0.1], autopct="%.2f")
plt.show()
```
